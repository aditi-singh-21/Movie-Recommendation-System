```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
df = pd.read_csv("final_dta.csv")
```

```python
stope_words_set = set(stopwords.words("english"))
from nltk.tokenize import word_tokenize

def remove_stop_words(summary):
    flag = 0
    summary_word_token = word_tokenize(summary)
    updated_summary = ""
    for w in summary_word_token:
        if w not in stope_words_set:
            if flag == 1:
                updated_summary = updated_summary + " " + w
            else:
                updated_summary = w
                flag = flag +1
    #print(updated_summary)
    return updated_summary

summary = df['Summary']
summary =list(summary)
update_summary_list =[]
for item in summary:
    update_summary_list.append(remove_stop_words(item))


    
    
```

```python
df['Summary'] = update_summary_list
```
