```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_list.csv")
```

```python
df.info()
```

```python
df['Year'].unique()
```

```python
plt.bar(df['Year'].unique() , df['Year'].value_counts() ,color= 'black')
plt.xlabel("Year")
plt.ylabel("Number of movies")
plt.title("Number of movies VS Year")
plt.show()
```

```python
reviews = list(df['User Reviews'])
name = list(df['Movie Name'])

df_name_reviews = pd.DataFrame({'User Reviews' : reviews , 'Movie Name' : name})
df_name_reviews = df_name_reviews.sort_values(by='User Reviews',ascending=False)
plt.barh(df_name_reviews['Movie Name'].iloc[0:10], df_name_reviews['User Reviews'].iloc[0:10],color= 'black')
plt.xlabel("User Ratings")
plt.ylabel("Movie Name")
plt.title("Movie vs User Rating")
plt.show()
```

```python
reviews = list(df['User Reviews'])
name = list(df['Movie Name'])

df_name_reviews = pd.DataFrame({'User Reviews' : reviews , 'Movie Name' : name})
df_name_reviews = df_name_reviews.sort_values(by='User Reviews')
#plt.barh(df_name_reviews['Movie Name'].iloc[0:10], df_name_reviews['User Reviews'].iloc[0:10],color= 'black')
plt.barh(df_name_reviews['Movie Name'].iloc[0:10], df_name_reviews['User Reviews'].iloc[0:10],color= 'black')
plt.xlabel("User Ratings")
plt.ylabel("Movie Name")
plt.title("Movie vs User Rating")
plt.show()
```

```python
rating = list(df['Rating'])
name = list(df['Movie Name'])

df_name_rating = pd.DataFrame({'Rating' : rating , 'Movie Name' : name})
df_name_rating = df_name_rating.sort_values(by='Rating',ascending=False)
plt.bar(df_name_rating['Movie Name'].iloc[0:10], df_name_rating['Rating'].iloc[0:10],color= 'black')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2,y[i], ha = 'center',bbox = dict(facecolor = 'white', alpha = .5))
rating = list(df_name_rating['Rating'])[0:10]
name = list(df_name_rating['Movie Name'])[0:10]
addlabels(name,rating)
plt.xlabel("Rating")
plt.xticks(rotation=90)
plt.ylabel("Movie")
plt.title("Number of movies VS Year")
plt.show()
```
