```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final_dta.csv")
```

```python
df.info()
```

```python
df['Year'].unique()
```

```python
df['Year'].value_counts()
```

```python
plt.bar(df['Year'].unique() , df['Year'].value_counts() ,color= 'black')
plt.xlabel("Year")
plt.ylabel("Number of movies")
plt.title("Number of movies VS Year")
plt.show()
```

```python
rating = list(df['Rating'])
name = list(df['Movie Name'])

df_name_rating = pd.DataFrame({'Rating' : rating , 'Movie Name' : name})
df_name_rating = df_name_rating.sort_values(by='Rating',ascending=False)
plt.bar(df_name_rating['Movie Name'].iloc[0:10], df_name_rating['Rating'].iloc[0:10],color= 'black')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2,y[i], ha = 'center',bbox = dict(facecolor = 'white', alpha = .5))
rating = list(df_name_rating['Rating'])[0:10]
name = list(df_name_rating['Movie Name'])[0:10]
addlabels(name,rating)
plt.xlabel("Rating")
plt.xticks(rotation=90)
plt.ylabel("Movie")
plt.title(" Top 250 movies")
plt.show()
```

```python
genre_count = list(df['Genre'].value_counts())[0:7]
genre_name = ['Action' ,'Comedy' ,'Drama' ,'Crime' ,'Biography' ,'Adventure','Mystery']
plt.pie(genre_count,labels=genre_name,autopct='%1.2f%%',textprops={'fontsize':6},colors=sns.color_palette(palette='binary'))
plt.show()
```

```python
dir_value = list(df['Director'].value_counts())[:10]
dir_name = ['Mani Ratnam','Anurag Kashyap','Vetrimaaran','Satyajit Ray','Rajkumar Hirani','Gautham Vasudev Menon','Jeethu Joseph','Hrishikesh Mukherjee','Karthik Subbaraj','S. Shankar']
plt.bar(dir_name, dir_value,color= 'black')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2,y[i], ha = 'center',bbox = dict(facecolor = 'white', alpha = .5))
addlabels(dir_name,dir_value)
plt.xlabel("Director")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.title("Count of Director")
plt.show()
```
